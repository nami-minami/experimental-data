```python
import numpy as np 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 

blanc = 0.003
Abs = np.array([0.003, 0.119, 0.239, 0.377, 0.475, 0.706]).reshape(-1,1) - blanc
mass = np.array([0, 5, 10, 20, 30, 40]).reshape(-1,1)

model = LinearRegression(fit_intercept=False)
model.fit(Abs, mass)
mass_pred = model.predict(Abs)

fig, ax = plt.subplots()
ax.set_xlim(0,40)
ax.set_ylim(0,0.73)

ax.scatter(mass, Abs, s=10)
ax.plot(mass_pred, Abs)
ax.text(25, 0.15, f"score = {model.score(Abs, mass):.3f}")
ax.text(10, 0.6, f"y = {model.coef_.item():.4f}x")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def alpha(pH):
    K1 = 1.1e-2
    K2 = 7.5e-8
    K3 = 4.8e-13
    H = 10**-pH 
    a0 = H**3/(H**3 + K1*H**2 + K1*K2*H + K1*K2*K3)
    a1 = K1*H**2/(H**3 + K1*H**2 + K1*K2*H + K1*K2*K3)
    a2 = K1*K2*H/(H**3 + K1*H**2 + K1*K2*H + K1*K2*K3)
    a3 = K1*K2*K3/(H**3 + K1*H**2 + K1*K2*H + K1*K2*K3)
    return a0,a1,a2,a3

pH = np.arange(0,14.1,0.1)
a0, a1, a2, a3 = alpha(pH)

plt.xlim(0,14)
plt.ylim(0,1)
plt.tick_params(direction = "in")

plt.plot(pH,a0)
plt.plot(pH,a1)
plt.plot(pH,a2)
plt.plot(pH,a3)
plt.savefig("buffer.pdf")
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 


pH = np.arange(0,14.1,0.1)
H = 10**-pH
D = 720/(H/9e-6 + 1 + 2e-10/H)
df = pd.DataFrame()
df["pH"] = pH
df["D"] = D
df.to_csv("pHvslogD.csv")

plt.xlabel("pH")
plt.xlim(0,14)
plt.ylabel("logD")
plt.ylim(-2.4, 3)
plt.tick_params(direction="in")
plt.scatter(pH, np.log10(D), s=5)
plt.savefig("pHvslogD.jpg")
```

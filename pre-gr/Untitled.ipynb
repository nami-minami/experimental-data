```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression

conc = np.array([0,2,5,10]).reshape(-1,1)
abs = np.array([0, 0.277, 0.705, 0.961]).reshape(-1,1)
sample_abs = np.array([1,2])

model = LinearRegression(fit_intercept=False).fit(conc,abs)
pred_abs = model.predict(np.arange(0,20).reshape(-1,1))

plt.scatter(conc,abs)
plt.plot(np.arange(0,20),pred_abs)
print(model.coef_, model.intercept_, model.score(conc,abs))
```

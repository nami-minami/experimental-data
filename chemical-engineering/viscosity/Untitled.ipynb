```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df = pd.read_csv("main.csv")
df.columns = ['乾球温度[℃]', '湿球温度[℃]', '質量[g]', '時間[min]']
df['含水率w'] = (df['質量[g]']-26.0137)/26.0137
df['含水率の変化量'] = -df['含水率w'].diff()
df
```

```python
plt.scatter(df['時間[min]'], df['含水率w'])
```

```python
plt.plot(df['時間[min]'], df['含水率の変化量'])
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 


df = pd.read_excel("粘度測定.xlsx", sheet_name="粘度", header=1, dtype='float32')
df
```

```python
conc = df['Unnamed: 0']
den = df['Unnamed: 5']
vis = df['実験値']


fig,ax = plt.subplots(1,2, figsize=(10,3), sharex=True, dpi=300)
ax[0].set_xlabel(r"Concentration")
ax[0].set_ylabel(r"Density")
ax[0].set_xticks(conc)
ax[0].set_xlim(0,100)
ax[0].set_ylim(0,1200)
ax[1].set_xlabel(r"Concentration")
ax[1].set_ylabel(r"Viscosity")
ax[1].set_ylim(0,1.8)
ax[1].set_yticks(np.arange(0,1.9,0.3))

ax[0].scatter(conc,den,s=7)
ax[1].scatter(conc,vis,s=7)

plt.savefig("graph.pdf")
```

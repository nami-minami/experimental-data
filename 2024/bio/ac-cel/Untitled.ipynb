```python
import numpy as np 
import scipy as sp
import pandas as pd 
import matplotlib.pyplot as plt
from IPython.display import display

df_dic = pd.read_excel("ir.xlsx", sheet_name=None, header=0, names=["spectrum", "abs"])
for df in df_dic.values(): 
    display(df)

print(df_dic.keys())
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,8), dpi=300)
labels = ["cellulose", "acetyl cellulose", "cellulose", "acetyl cellulose"]
index=0

for df_item in df_dic.items():
    if df_item[0] == 'ろ紙.0' or df_item[0] == '6班 ろし酢酸セルロース.0':
        ax[0].plot(df_item[1]['spectrum'], df_item[1]['abs'], label=labels[index])
    else:
        ax[1].plot(df_item[1]['spectrum'], df_item[1]['abs'], label=labels[index])
    index+=1

for i in range(2):
    ax[i].invert_xaxis()
    ax[i].legend(loc="upper left")
    ax[i].plot((1725,1725),(0,0.1), color="black")
    ax[i].plot((1170,1170),(0,0.1), color="black")

ax[0].set_title("IR spectrum of cotton", y=-0.1, fontsize="xx-large")
ax[1].set_title("IR spectrum of reed", y=-0.1, fontsize="xx-large")
```

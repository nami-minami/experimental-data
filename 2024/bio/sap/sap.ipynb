# 吸水性ポリマー

## 吸水率

```python
import pandas as pd
from IPython.display import display

df_dic = {
    "H2O": pd.DataFrame(
        index=[ 
            "CMC SAP",  
            "PANa",  
            "Empty" 
        ],
        data={
            "試料量": [
                0.13,
                0.10, 
                0
            ],
            "吸水前全質量": [
                0.51, 
                0.51, 
                0.36
            ],
            "吸水後全質量": [
                29.82,
                41.99,
                2.13,
            ],
        }
    ),
    "NaCl": pd.DataFrame(
        index=[
            "CMC SAP", 
            "PANa", 
            "Empty"
        ],   
        data={
            "試料量": [
                0.11,
                0.10,
                0
            ],
            "吸水前全質量": [
                0.49, 
                0.48, 
                0.38
            ], 
            "吸水後全質量": [
                8.61,
                6.81,
                2.32
            ],
        } 
    ),
    "CaCl2": pd.DataFrame(
        index=[
            "CMC SAP", 
            "PANa", 
            "Empty"
        ],   
        data={
            "試料量": [
                0.10,
                0.14,
                0
            ],
            "吸水前全質量": [
                0.46, 
                0.53, 
                0.37
            ], 
            "吸水後全質量": [
                5.35,
                2.51,
                1.49
            ],  
        } 
    )
}
```

```python
for df in df_dic.values():
    df['吸水量'] = (df['吸水後全質量'] - df.at['Empty', '吸水後全質量'] - df['試料量']) / df['試料量']
    display(df.round(2))

with pd.ExcelWriter("吸水率.xlsx") as writer:
    df_dic['H2O'].to_excel(writer, sheet_name="H2O")
    df_dic['NaCl'].to_excel(writer, sheet_name="NaCl")
    df_dic['CaCl2'].to_excel(writer, sheet_name='CaCl2')
```
